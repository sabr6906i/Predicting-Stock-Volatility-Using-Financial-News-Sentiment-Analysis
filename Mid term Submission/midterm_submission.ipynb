#Task-1

```python
import string
import requests
import time
import pandas as pd
import yfinance as yf
from bs4 import BeautifulSoup

lst = list(range(1,20))
df = pd.DataFrame(index=lst, columns=["Source", "Headline", "PubDate"])
url = "https://feeds.finance.yahoo.com/rss/2.0/headline?s=ADMA"


print(df)
headers = {
    "User-Agent": "Mozilla/5.0",
    "Accept": "application/rss+xml,application/xml"
}

time.sleep(3)
resp = requests.get(url, headers=headers)

soup = BeautifulSoup(resp.content, "xml")

items = soup.find_all("item")
for k,i in zip(items, df.index):
    df.loc[i, 'PubDate'] = k.pubDate.text
    df.loc[i, 'Headline'] = k.description.text
    df.loc[i, 'Source'] = k.link.text
df.iloc[6:30,:]
```

```python
df.to_csv('News_raw.csv')
```

```python
df['PubDate'] = df['PubDate'].str.split(' ').str[1:4].str.join(" ")
df['PubDate'] = pd.to_datetime(df['PubDate'],format="%d %b %Y")
df['PubDate'] = pd.to_datetime(df['PubDate'],format="%Y %b %d")
df
```

```python
df['Headline Length'] = 9
for i in df.index:
    df.loc[i, 'Headline Length'] = len(df.loc[i, 'Headline'])

df.to_csv('news_cleaned.csv')
```
